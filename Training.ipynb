## Training Loop

```python
agents_per_class = {}
for i in tq.tqdm(range(len(classes))):
    classe = classes[i]
    print("Classe "+str(classe)+"...")
    agents_per_class[classe] = Agent(classe, alpha=0.2, num_episodes=1, load=False, model_name='vgg16')
    agents_per_class[classe].train(datasets_per_class[classe])
    del agents_per_class[classe]
    torch.cuda.empty_cache()
```
